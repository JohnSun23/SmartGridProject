```python
tag = 'NAE-21_MAYER.N2-2.DX-030.SETPOINTS.SETPOINTS.HSAT-STPT.HIGH_SUPPLY_AIR_TEMP_SETPOINT'
print 'tag: %s'%tag
df = get_tag_series(tag)
plot_daily_series(df.head(n=5),title='original')
plot_daily_series(df_svd.head(n=5),title='reconstructed: svd 15')
```

```python
# Matrix of signal: days * minutes
X = df.values
# (number of days, minutes per day)
(n,m) = X.shape        
# If there are only <n> days, we can't run SVD with k > n
k = min(k,n)
# Subtract off the mean from every row
M = X.mean(axis=0)
X2 = X - M
# SVD
try:
    U,s,Vh = svd(X2, full_matrices=False)
except (np.linalg.linalg.LinAlgError, ValueError) as _:
    U = np.zeros((n,n))
    s = np.zeros((n,))
    Vh = np.zeros((n,m))
U = U[:,:k]
s = s[:k]
Vh = Vh[:k,:]
```

```python
pylab.rcParams['figure.figsize'] = (12.0, 10.0)
for i in random.sample(range(len(X)),4):
    plot(X[i,:],color='black',label='original')

#y = M
#for i in range(15):
#    y = y + U[0,i]*s[i]*Vh[i,:]
#plot(y,color='red')


plot(M,color='red',label='mean',linewidth=3)
i = 0
plot(M+U[0,i]*s[i]*Vh[i,:],color='green',label='svd component 0',linewidth=3)
i = 1
plot(M+U[0,i]*s[i]*Vh[i,:],color='blue',label='svd component 1',linewidth=3)

legend()
show()
```
