### Overview ###
For all sensors, we've calculated the space/errors (default MSE) for multiple compression methods/complexity hyperparameters.  

In this notebook, we generate single tradeoff curves per sensor combining all compression methods/hyperparameters.  

We then reduce the dimensionality of these curves, and use these to group sensors.    

### Load Space/Error ###
The dataframe "se" gives the space/error tradeoff for each sensor/compression method.   

* __tag__ string sensor name, eg 'NAE-10_BIOMED-LIB.N2-2.AC-15.DX001.OA-T.OUTSIDE_AIR_TEMPERATURE'
* __compressor__ string compression method, eg 'svd'
* __space_err__ b64 serialized (see `<project root>/serialize_tag_date.py/b64_...`) numpy array giving space/error tradeoff
  * first half of numpy array gives monotonically-increasing space taken per hyperparameter k (the index into array)
  * second half of array gives MSE (or other error metric used) corresponding to space in first half
  * the function get_space_mse b64-deserializes and splits these
  
se2 is se, for a single reasonable outside temperature sensor.  

```python
se_path = os.path.join(os.environ['compression_data_dir'],'all_space_err.txt')
se = pd.read_csv(se_path,names=['tag','compressor','space_err'],sep='^')

tag = 'NAE-10_BIOMED-LIB.N2-2.AC-15.DX001.OA-T.OUTSIDE_AIR_TEMPERATURE'
se2 = se[se['tag'] == tag]
```

### Compute/Plot Space/Error Tradeoff ###
We know space/error tradeoff for each compression method/sensor, but what about for a sensor, combining compression methods? 
`get_space_mse_tradeoff` combines compression methods, and a dataframe representing the combination.  See help doc for more details
on what `get_space_mse_tradeoff` does.  
We then plot a color-coded version of this tradeoff.  

```python
tradeoff = get_space_mse_tradeoff(se2)
plot_space_mse_tradeoff(tradeoff)
```

### Reduce Tradeoff Dimensionality ###
The space/error tradeoff curve includes far too many points: for SVD, the number of eigenvectors could range from 
0 up to the number of days per sensor (hundreds). 

We cluster the space used in log-space, and select a representative sample of this tradeoff error.  

Compare this reduced-dimensional curve to the curve above.  The information we care about seems to be captured: KMeans in log-space works quite well.  


```python
spaces2 = choose_space_binning(tradeoff,k=10)
tradeoff2 = apply_space_binning(spaces2,tradeoff)
```

```python
plot_space_mse_tradeoff(tradeoff)
title('tradeoff, selecting %d representative space/errs')
plot_space_mse_tradeoff(tradeoff2)
show()
```

### Cluster Sensors using Tradeoff ###
For each sensor, we map the reduced-dimensional tradeoff into log-space.  This gives a feature-vector for each sensor.  

We then cluster sensors in this space.  

```python
%time all_errs = se.groupby('tag',as_index=False).apply(lambda x:pd.Series(space_err_features(spaces2,x)))
```

```python
# Remove flat signals: these go in the first group
all_errs2 = all_errs.replace([np.inf,-np.inf,np.nan],np.nan)
all_errs2[all_errs2 <= 0] = np.nan
all_errs2 = all_errs2.dropna()
all_errs2 = np.log(all_errs2)
all_errs2 = all_errs2[range(7)]
m = len(all_errs2)
```

```python
# Cluster the remaining signals
k = 100
model = KMeans(n_clusters=k)
C = model.fit_predict(all_errs2)
cluster_dist = dict(FreqDist(C.tolist()).items())
cluster_dist = array([cluster_dist[c] for c in range(k)])
```

# Relating Clusters to Tag Types #
The type of a tag can be extracted from the last part of the name, eg air flow/outside temperature/...

Above, we clustered sensors based on their __values__, with the hope that this grouping of sensors 
would make sense to an administrator.  Here we demonstrate a number of strong relationships.  In fact, 
we argue that the grouping by sensor values may provide a __more meaningful__ grouping than grouping 
by sensor types.  

The two approaches we'll take are:

* Demonstrate a strong statistical relationship between cluster assignments and tag types.  This shows the clustering is capturing deep information about the sensors.  
* Show that the clustering can be used to group tag types.  We demonstrate qualitatively that this grouping is meaningful.  

### Statistical Clusters and Tag Types ###
We show there is significant mutual information between the signal-value based clustering and tag types, which are extracted from the tag name.  

Mutual information is difficult to compute, since the joint distribution is sparse, and zero-values result in singularities.  

Since the relationship between clusters and tag types is strong, we can use a very simple test:
Assume we have sensors enumerated $S = \{1,...m\}$.  Let $\tau_t$ for $t \in S$ be the tag type of sensor $t$, and $c_t$ be the cluster label for tag $t$.  

Then $J_t = (\tau_t,c_t)$ is a single sample from the tag type/cluster label distribution.  Let $u(J) = |\{J_t \mid t \in S\}|$ be the number of distinct elements in the sample from the joint distribution.  Given the null, that $\tau_t$ and $c_t$ have no mutual information, we would expect $u(J)$ to be the same, whether $\tau_t$ is permuted or not.  We show this is clearly not the case: there are far fewer distinct samples in the joint.  

```python
tag_list = all_errs2.index
# Select most common tag types, such that >=90% of tags are accounted for
#tag types, per tag
tag_types = array(map(tag_type,tag_list))
# counts of tag types
tt_dist = FreqDist(tag_types.tolist())
```

```python
# counts up to ith tag type
counts = np.array([c for _,c in tt_dist.items()]) 
counts = np.cumsum(counts/float(counts.sum()))
# tt_dist[:i] tag types account for 90% of tags
n = np.nonzero(counts >= 0.9)[0][0] + 1
# common tag types that make up 90% of tags
common_tt = [tag for tag,c in tt_dist.items()[:n]]+['uncommon']
common_tt_hash = dict( ((t,i) for i,t in enumerate(common_tt)) )
```

```python
# bitvector of tag type, over all tags
tts = np.zeros((m,n+1)) #+1 for "uncommon tag"
# index of tag type (up to n+1), over all tags
tts2 = np.zeros((m,)) #+1 for "uncommon tag"
for i,tag_type in enumerate(tag_types):
    if tag_type in common_tt_hash.keys():
        j = common_tt_hash[tag_type]
    else:
        j = -1 #uncommon
    tts[i,j] = 1
    tts2[i] = j
tts = pd.DataFrame(tts)
tts['C'] = C
tt_df = tts.groupby('C').sum()
tt_df = tt_df / tt_df.sum(axis=0)
```

```python
tts2_shuffled = tts2.copy()
duplicates = len(set(zip(tts2,C)))
shuffled_duplicates = []
for _ in range(1000):
    random.shuffle(tts2_shuffled)
    shuffled_duplicates.append(  len(set(zip(tts2_shuffled,C)))  )
shuffled_duplicates = np.array(shuffled_duplicates)
```

```python
hist(shuffled_duplicates,bins=40,normed=True,label='shuffled unique',edgecolor='none')
vlines([duplicates],0,0.2,label='actual unique',color='red')
legend()
show()
```

### Improved Sensor Grouping ###
We show that sensor groupings give informative groupings of tag types.  `ttdf[i,j]` gives the proportion of tags of 
the $j^th$ tag type that is in the $i^th$ cluster.  If we group tags by this distribution over sensor-value-clusters, 
reducing from 61 tag types to 15 groups, we capture additional information about the signals.  

An example of such a group of tag types is the following

    DAMPER_POSITION
    REHEAT_VALVE_COMMAND
    HEATING_COMMAND
    COOLING_COMMAND
    Reheat_Valve_Command
    HEATING_VALVE_COMMAND
    COOLING_COIL_VALVE_COMMAND
    Frequency_Output
    
This grouping collected COMMAND-type tags into one group. 

Another clustering grouped SUPPLY_AIR_FLOW and SUPPLY_AIR_FLOW_SETPOINT tags together.  For example:

`NAE-06_EBU3B.N2-2.RM-4217.VMA158.SUP-FLOW.SUPPLY_AIR_FLOW`

and

`NAE-06_EBU3B.N2-2.RM-4217.VMA158.SUPFLO-SP.SUPPLY_AIR_FLOW_SETPOINT`

were in the same cluster.  As shown by the plots below, these signals have very similar properties for the same room, and should be considered together.  

```python
tt_labels = KMeans(n_clusters=15).fit_predict(tt_df.values.T)
```

```python
# Print top 3 cluster tag types
for c in map(itemgetter(0),FreqDist(tt_labels.tolist()).items()[:3]):
    for i in np.nonzero(tt_labels==c)[0]:
        print common_tt[i]
    print ''
```
