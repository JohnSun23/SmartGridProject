```python
errs = []
x = df.values[0,:]
coeffs = pywt.wavedec(x,'haar')
zero_coeffs = pywt.wavedec(np.zeros((1440,)),'haar')

#####################33
coeff_indices = [(c,(l,i)) for l,cs in enumerate(coeffs) for i,c in enumerate(cs)]
coeff_indices = pd.Series(*zip(*coeff_indices))
top_coeff_indices = sorted(coeff_indices.index, key=lambda i:abs(coeff_indices[i]))[-20:]
for (l,i) in top_coeff_indices:
    zero_coeffs[l][i] = coeffs[l][i]
x2 = pywt.waverec(zero_coeffs,'haar')
##############################
#coeffs3 = fill_coeffs(coeffs[:5],coeff_lens)
#x3 = pywt.waverec(coeffs3,wavelet)
coeff_lens = map(len,coeffs)
errs.append(('haar', norm(x - x2)))
#print norm(x[:1440] - x3[:1440])
plot(x,color='black',label='original')
plot(x2,color='red',label='wavelets by err')
#plot(x3[:1440],color='blue',label='wavelets by resolution')
legend(bbox_to_anchor=(1.6, 0.8))
show()

print sorted(errs,key=itemgetter(1))
```
