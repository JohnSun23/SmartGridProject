# Filter Out Trivially Compressed Tag/Days #

```python
# baseline is: a constant for the entire sensor, across all days
n = float(len(df_compressed)) #Original number of tag/days
print 'Originally %d tag/days'%n
errs = df_compressed.xs('error',axis=1,level=1)

# Remove PowerAnalytics tags
power_analytics_tags = set(i for i,tag in enumerate(context['tag_list']) if 'PowerAnalytics' in tag)
power_analytics_filter = np.array([(i in power_analytics_tags) for i in errs.index.get_level_values(0)])
errs = errs[~power_analytics_filter]
print 'After removing PowerAnalytics data: %d, %.3f of total'%(len(errs),len(errs)/n)

#Remove easily compressed tag-days
errs = errs[errs['constant_tag'] > 0] 
print 'After filtering out constants over entire tag: %d, %.3f of total'%(len(errs),len(errs)/n)
errs = errs[errs['constant_tag_day'] > 0]
print 'After filtering out constants over each tag/date: %d, %.3f of total'%(len(errs),len(errs)/n)
errs = errs[errs['haar_wavelet_4'] > 0]
print 'After filtering out cases where haar perfectly compresses: %d, %.3f of total'%(len(errs),len(errs)/n)

# Scale errors relative to constant-per-day
columns = set(errs.columns)
other_columns = list(columns - set(['constant_tag_day']))

for c in other_columns:
    errs[c] = errs[c] / errs['constant_tag_day']
    
errs['constant_tag_day'] = 1

# Show summary of scaled errors
errs.describe()
```

# Split/Visualize Series By Compression Error #

```python
def filter_df_date(df, date_index, context):
    ''' Given a DataFrame with tag/date indices, and floating point values for each day,
        select a particular tag/date.  Return the new DataFrame '''
    date_str = context['date_list'][date_index].replace('_','/')
    date_filter = (df.index.get_level_values(1) == date_str)
    return df[date_filter]

def error_filter(errs, compressor, error_range):
    ''' Return an array of booleans, whether each tag-date satisfies:
        error for the given compressor is in the given (min,max) err_range '''
    return (errs[compressor] > error_range[0]) & (errs[compressor] < error_range[1])
```

```python
def plot_tag_day_reconstructed(tag_index, date_index, compressors):
    tag = context['tag_list'][tag_index]
    print 'tag: %s'%tag
    df = get_tag_series(tag)
    plot_daily_series(df, title='samples from sensor')
    df_day = filter_df_date(df, date_index, context)
    plot_daily_series(df_day,samples=None,title='sensor for chosen day')
    for compressor in compressors:
        outlier_reconstructed = decompress_df(df_compressed[(df_compressed.index.get_level_values(0) == tag_index) & 
                                                   (df_compressed.index.get_level_values(1) == date_index)],
                                                  context, compressor)
        plot_daily_series(outlier_reconstructed,samples=None, title='reconstructed day using %s'%compressor)

tag_index,date_index = random.choice(haar_does_well)

plot_tag_day_reconstructed(tag_index,date_index, ['haar_wavelet_4','svd_2'])
```

```python
# Create a dict mapping tag -> [selected dates]
dist = defaultdict(list)
for t,d in selected:
    dist[t].append(d)
```

###  Select Outliers ###

```python
# Select out tags that have few matching cases
# Add the days for these tags to the list of outliers
outliers = []
for t,ds in dist.iteritems():
    if len(ds) < 5:
        for d in ds:
            outliers.append((t,d))
```

```python
# Select outliers in the time range of the Sept 8 blackout
blackout_outliers = [(t,d) for t,d in outliers if d > 0 and d < 400]

# Group outlier tags by type
outlier_tags = set([t for t,d in blackout_outliers])
partitioned = defaultdict(set)
remaining_tags = outlier_tags
for tag_type in ['heating', 'damper', 'air', 'freq', 'vol', 'power', 'occupied', '^ch', 'vll', 'kwh',
                  'kva', 'temp','warm','valve_command','current','kw','vln','box_mode','cooling','pf_lag']:
    matching_tags = set(t for t in remaining_tags if 
                        not re.search(tag_type,context['tag_list'][t],re.IGNORECASE) is None)
    remaining_tags -= matching_tags
    partitioned[tag_type] = matching_tags
partitioned['other'] = remaining_tags

# Select outliers of the given tag type
tag_types = ['freq','vol','power','vll','kwh','kva','kw','vln','pf_lag'] #electricity-related tags
blackout_outliers = [(t,d) for t,d in blackout_outliers 
                     if t in set.union(*[partitioned[tag_type] for tag_type in tag_types])]
```

```python
# Plot a randomly chosen outlier, along with random
#  other days from the same sensor
tag_index, date_index = random.choice(blackout_outliers)
tag_str = context['tag_list'][tag_index]
date_str = context['date_list'][date_index]
print tag_str, date_str
df = get_tag_series(tag_str)
outlier_reconstructed = decompress_df(df_compressed[(df_compressed.index.get_level_values(0) == tag_index) & 
                                               (df_compressed.index.get_level_values(1) == date_index)],
                                      context, 'haar_wavelet_4')
plot_daily_series(df,samples=10)
df_outlier = filter_df_date(df,date_index,context)
plot_daily_series(df_outlier,samples=None)
plot_daily_series(outlier_reconstructed,samples=None)
```
