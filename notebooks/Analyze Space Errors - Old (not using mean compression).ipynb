```python
def get_optimal_shift_error(x0, x1, max_shift):
    ''' Get optimal shift of pandas series x1, so it aligns well with x0 '''
    shift_errs = []
    for shift in range(-max_shift,max_shift+1,1):
        shift_errs.append((  shift, norm(x0 - x1.shift(periods=shift).fillna(x1.mean()))  ))
    return min(shift_errs, key=itemgetter(1))

def load_compressed_tag_errs(err_file_basename):
    all_vars = {'df': pd.load(err_file_basename+'.pandas') }
    with open(err_file_basename+'.pkl','r') as f:
        all_vars.update(pickle.load(f))
    return all_vars

def sample_df(df, samples):
    rows = random.sample(df.index, samples)
    return df.ix[rows]

def plot_daily_series(df, samples=5, title=None):
    ''' Given a dataframe containing daily series for sensors, 
        randomly sample <samples> days, and plot the daily variation for those days'''
    if not samples is None:
        df = sample_df(df, samples)
    clf()
    X = df.copy()
    # Find the minimum and maximum
    ymax = X.max().max()
    ymin = X.min().min()
    yrange = ymax - ymin
    fig = figure(figsize=(15,5))
    ax = fig.add_subplot(221)
    fig.suptitle('yrange: %.5f'%yrange)
    if yrange == 0:
        yrange = 1
    #Add a small amount of random noise, so we can see overlapping lines
    X = X + np.random.randn(*X.shape)*0.01*abs(yrange) 
    X.index = X.index.droplevel(0)
    X.T.plot(ax=ax, grid=False, legend=False, title=title)
    ax.legend(loc='center left', bbox_to_anchor=(1,0.5), fancybox=True, shadow=True)
    # Set reasonable x/y limits.  y slightly above/below the largest values
    xlim([0,1440])
    if yrange == 0:
        ymax, ymin = (-1,1)
    else:
        ymax += 0.1*yrange
        ymin -= 0.1*yrange
    ylim([ymin, ymax])
    show()
```
