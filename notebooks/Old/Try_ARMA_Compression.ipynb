```python
def get_optimal_shift_error(x0, x1, max_shift):
    ''' Get optimal shift of pandas series x1, so it aligns well with x0 '''
    shift_errs = []
    for shift in range(-max_shift,max_shift+1,1):
        shift_errs.append((  shift, norm(x0 - x1.shift(periods=shift).fillna(x1.mean()))  ))
    return min(shift_errs, key=itemgetter(1))

def lms_fit(X,Y,p,u=1):
    n = len(X)
    h = array([0]*p)
    errs = []
    for i in range(p,n-1):
        x = array(X[i-p:i])
        err = Y[i] - (h*x).sum()
        h = h + u*err*x/((x**2).sum())
        errs.append(err)
    return h,np.array(errs)

def lms_predict(X_begin,n,p,h):
    Y = np.zeros(shape=(n,))
    Y[:p] = X_begin
    for i in range(p,n-1):
        Y[i] = (h*Y[i-p:i]).sum()
    return Y

def lms_compress(X,n,p,h,epsilon):
    interpolated = np.ones(shape=(n,))
    Y = np.zeros(shape=(n,))
    Y[:p] = X[:p]
    interpolated[:p] = 0
    errs = []
    for i in range(p,n):
        Y[i] = (h*Y[i-p:i]).sum()
        err = abs(Y[i] - X[i])
        if err > epsilon:
            Y[i] = X[i]
            interpolated[i] = 0
            errs.append(0)
        else:
            errs.append(err)
    return Y,interpolated,np.array(errs)
```
