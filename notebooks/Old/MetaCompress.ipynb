```python
def pvr(se):
    ''' Given a compression ration of 100, what percentage variance reduction can we get?'''
    tradeoff = get_space_mse_tradeoff(se)
    baseline = tradeoff[tradeoff['compressor'] == 'constant_tag']['error'].values[0]
    err = tradeoff[tradeoff['compression_ratio'] > 100]['error'].min()
    return (err**2 - baseline**2)/baseline**2
```

# Build Meta-Compressor #

```python
def choose_compressor(tag_err):
    ''' Given a dictionary mapping <compression method> -> error, 
        choose a compression method '''
    if tag_err['constant_tag'] == 0:
        compressor = 'constant_tag'
    elif tag_err['haar_wavelet_4'] < tag_err['svd_2']:
        compressor = 'haar_wavelet_4'
    else:
        compressor = 'svd_2'
    return compressor

def meta_compress(df_compressed):
    ''' Given a dataframe storing a compressed representation of the original data for each tag/day
        select out the error stored, and choose a compression method based on this error '''
    tag_err = df_compressed.xs('error',axis=1,level=1).mean()
    compressor = choose_compressor(tag_err)
    return df_compressed[compressor]

# Also needs global variables compressors, context, and meta_df
def meta_decompress(tag_index):
    ''' Select and compress by the best compression method, for each sensor '''
    compressor = compressors[  compressor_choices.ix[tag_index]  ]
    tag = context['tag_list'][tag_index]
    aggregate = context['aggregates'][compressor.name()][tag]
    aggregate = b64_decode_series(aggregate)
    meta_df_tag = meta_df[meta_df.index.get_level_values(0) == tag_index]
    return compressor.decompress(aggregate, meta_df_tag)
```

### Meta-Compress ###
Collect space/error statistics for each sensor, based on per sensor-day data

Choose compressors based on this information, and compress the data

```python
# Compute mean space/error for each tag
tag_stats = df_compressed.groupby(df_compressed.index.get_level_values(0)) \
    [[(a,b) for a,b in df_compressed.columns if b != 'compressed']].mean()
# Get the name of the chosen compressor for each tag
compressor_choices = tag_stats.xs('error',axis=1,level=1).apply(choose_compressor,axis=1)
# Only store the chosen compressed representation for each tag
df2 = df_compressed.head(n=5000)
meta_df = df2.groupby(df2.index.get_level_values(0)).apply(meta_compress)
```

### Load Space/Error Results ###

```python
def get_space_mse(se):
    ''' Deserialize 64-bit-encoded space/error.  return (space,error)'''
    se = b64_decode_series(se)
    n = len(se)
    return se[:n/2],se[n/2:]
```

```python
def get_space_mse_tradeoff(se):
    ''' Given a part of the space vs. error dataframe'''
    spaces = []
    errs = []
    compressors = []
    for _,row in se.iterrows():
        s,e = get_space_mse(row['space_err'])
        c = [row['compressor']]*len(s)
        spaces += s.tolist()
        errs += e.tolist()
        compressors += c
    tradeoff = pd.DataFrame({'compression_ratio':spaces,'error':errs,
                         'compressor':compressors}).sort(columns=['compression_ratio'],ascending=False)
    l = 0
    while len(tradeoff) != l:
        l = len(tradeoff)
        decreased = tradeoff['error'].diff() < 0
        decreased[0] = True
        tradeoff = tradeoff[decreased]
    return tradeoff
```

# Correlate Compresses Well with Tag Name #

```python
def get_room(tag):
    '''Get the room number in the building for the tag.
       Definintely works well for EBU3B, not so sure about other buildings'''
    spl = re.split('[^A-Za-z0-9]',tag) #Split tag by non-alpha-numeric
    try:
        #The room number comes directly after 'RM
        rm_index = (i for i,s in enumerate(spl) if 'RM' in s).next()
        return spl[rm_index+1]
    except StopIteration:
        return None

def get_building(tag):
    '''Get the building for NAE tags.  Works well for all such NAE tags
       Returns 'None' if there was an error, or the tag is not an NAE'''
    if tag[0:3] != 'NAE':
        return None
    tag_remainder = tag.split('_',1)[1]
    return re.split(r'\W',tag_remainder)[0]

def correlate(in_tag):
    # bitvector of tags containing all strings in the list <in_tag>, case insensitive
    tag_filter = np.array([all([t in tag.lower() for t in in_tag]) for tag in tag_list])
    n = sum(tag_filter) #total number of tags matching
    dist = nltk.FreqDist(compressor_choices[tag_filter].values.tolist()).items()
    dist = pd.DataFrame(dist,columns=['compressor','proportion'])
    dist['proportion'] = dist['proportion'] / float(dist['proportion'].sum())
    return dist,n
```

# Split/Visualize Series By Compression Error #

```python
def filter_df_date(df, date_index, context):
    ''' Given a DataFrame with tag/date indices, and floating point values for each day,
        select a particular tag/date.  Return the new DataFrame '''
    date_str = context['date_list'][date_index].replace('_','/')
    date_filter = (df.index.get_level_values(1) == date_str)
    return df[date_filter]

def error_filter(errs, compressor, error_range):
    ''' Return an array of booleans, whether each tag-date satisfies:
        error for the given compressor is in the given (min,max) err_range '''
    return (errs[compressor] > error_range[0]) & (errs[compressor] < error_range[1])
```
