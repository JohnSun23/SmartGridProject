```python
def choose_compressor(tag_err):
    ''' Given a dictionary mapping <compression method> -> error, 
        choose a compression method '''
    if tag_err['constant_tag'] == 0:
        compressor = 'constant_tag'
    elif tag_err['haar_wavelet_4'] < tag_err['svd_2']:
        compressor = 'haar_wavelet_4'
    else:
        compressor = 'svd_2'
    return compressor

def meta_compress(df_compressed):
    ''' Given a dataframe storing a compressed representation of the original data for each tag/day
        select out the error stored, and choose a compression method based on this error '''
    tag_err = df_compressed.xs('error',axis=1,level=1).mean()
    compressor = choose_compressor(tag_err)
    return df_compressed[compressor]

# Also needs global variables compressors, context, and meta_df
def meta_decompress(tag_index):
    ''' Select and compress by the best compression method, for each sensor '''
    compressor = compressors[  compressor_choices.ix[tag_index]  ]
    tag = context['tag_list'][tag_index]
    aggregate = context['aggregates'][compressor.name()][tag]
    aggregate = b64_decode_series(aggregate)
    meta_df_tag = meta_df[meta_df.index.get_level_values(0) == tag_index]
    return compressor.decompress(aggregate, meta_df_tag)

def get_space_mse(se):
    ''' Deserialize 64-bit-encoded space/error.  return (space,error)'''
    se = b64_decode_series(se)
    n = len(se)
    return se[:n/2],se[n/2:]

def choose_space_mse(se):
    s,e = get_space_mse(se)
    i = random.randint(0,len(s)-1)
    return s[i],e[i]

def tag_space_error(se):
    ''' Given a sensor, output 3 arrays, sorted by the space array'''
    all_s = []
    all_e = []
    all_c = []
    for c in se['compressor']:
        s,e = get_space_mse(se[se['compressor'] == c]['space_err'].values[0])
        all_s += s.tolist()
        all_e += e.tolist()
        all_c += [c]*len(s)
    #all_s = np.array(all_s)
    #all_e = np.array(all_e)
    #all_c = np.array(all_c)
    #print len(all_s),len(all_e),len(all_c)
    df = pd.DataFrame({'compressor':all_c,'mse':all_e},index=all_s)
    df.index.name = 'space'
    df.sort(axis=0,inplace=True)
    #print df
    return df

def choose_space_mse(se):
    s,e = get_space_mse(se)
    i = random.randint(0,len(s)-1)
    return s[i],e[i]
```

```python
def np_first(x):
    ''' Given numpy 1D boolean array x, return first True value
        return None if there is no such value '''
    z = np.nonzero(x)[0]
    if len(z) > 0:
        return z[0]
    else:
        return None

def best_mse(se,compression_ratio):
    ''' Return space/compressor/mse with min error, such that compression ratio >= compression'''
    i = np_first(se.index.get_level_values(1) >= compression_ratio)
    if i is None:
        i = len(se)-1
    good_compression = se.ix[i:]
    i = good_compression['mse'].argmin()
    best = good_compression.ix[i].copy()
    compression_ratio = good_compression.index[i][1]
    baseline_mse = se[se['compressor'] == 'constant_tag']['mse']
    if len(baseline_mse) > 0:
        baseline_mse = baseline_mse.values[0]
    else:
        baseline_mse = np.NAN
    return pd.DataFrame(
                        {'mse':[best['mse']],
                         'mse_improvement':[baseline_mse/best['mse']],
                         'compression_ratio':[compression_ratio], 
                         'compressor':[best['compressor']]}
                        )
```
