# Reconstruct From Meta-Compressed Representation #
Load a meta-compressed representation of the sensor data.  Select sensors, and plot the reconstructed vs. original signals.

meta-compression chooses the best compressor among methods and hyperparameters, to balance a tradeoff of space and error.  

### Load meta-compressed data ###

```python
compression_dir = os.path.join(os.environ['SMART_GRID_DATA'],'compression')
context = load_compressed_tag_errs(os.path.join(compression_dir,'best_compressed'))
df_compressed = context['df']
tag_list = context['tag_list']
```

### Select a sensor ###

```python
#tag = random.choice(tag_list)
# Load best compressor choices
best_path = os.path.join(os.environ['compression_data_dir'],'best.pandas')
best = pd.load(best_path)
# Choos tag that's compressed well with svd, but not too well
tag = random.choice(best[(best['compressor'] == 'svd') & (best['k'] < 20) & (best['k'] > 15)].index)
# Choos tag that's compressed well with wavelet, but not too well
#tag = random.choice(best[(best['compressor'] == 'wavelet') & (best['k'] < 7) & (best['k'] > 7)].index)
```

### Decompress/Load Original ###

```python
df_meta = meta_decompress_tag(df_compressed,tag,context)
df = get_tag_series(tag)
best.ix[tag]
```

### Visualize Decompressed vs. Original ###

```python
# Plot original/reconstructed for some random dates
random_dates = random.sample(df_meta.index,5)
plot_daily_series(df.ix[random_dates],title='original')
plot_daily_series(df_meta.ix[random_dates],title='reconstructed')
```
